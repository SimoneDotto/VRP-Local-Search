```python
import os
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df_SA = pd.read_csv("Log_SA.log", sep=" ")
df_SA["Profit"] = -df_SA["Cost"]
df1_SA = df_SA[df_SA["Profit"] == df_SA.groupby("Instance")["Profit"].transform(max)].set_index("Instance").sort_index()

df_SAG = pd.read_csv("Log_SAG.log", sep=" ")
df_SAG["Profit"] = -df_SAG["Cost"]
df1_SAG = df_SAG[df_SAG["Profit"] == df_SAG.groupby("Instance")["Profit"].transform(max)].set_index("Instance").sort_index()

df_HC = pd.read_csv("Log_HC.log", sep=" ")
df_HC["Profit"] = -df_HC["Cost"]
df1_HC = df_HC[df_HC["Profit"] == df_HC.groupby("Instance")["Profit"].transform(max)].set_index("Instance").sort_index()

df_HCG = pd.read_csv("Log_HCG.log", sep=" ")
df_HCG["Profit"] = -df_HCG["Cost"]
df1_HCG = df_HCG[df_HCG["Profit"] == df_HCG.groupby("Instance")["Profit"].transform(max)].set_index("Instance").sort_index()

df_SD = pd.read_csv("Log_SD.log", sep=" ")
df_SD["Profit"] = -df_SD["Cost"]
df1_SD = df_SD[df_SD["Profit"] == df_SD.groupby("Instance")["Profit"].transform(max)].set_index("Instance").sort_index()

df_SDG = pd.read_csv("Log_SDG.log", sep=" ")
df_SDG["Profit"] = -df_SDG["Cost"]
df1_SDG = df_SDG[df_SDG["Profit"] == df_SDG.groupby("Instance")["Profit"].transform(max)].set_index("Instance").sort_index()

display(df1_HC)
display(df1_HCG)
display(df1_SD)
display(df1_SDG)
display(df1_SA)
display(df1_SAG)
```

```python
df_SA["Time"] = df_SA["Time"].apply(lambda x : float(x[:-1]))
df_HC["Time"] = df_HC["Time"].apply(lambda x : float(x[:-1]))
df_SD["Time"] = df_SD["Time"].apply(lambda x : float(x[:-1]))
df_SAG["Time"] = df_SAG["Time"].apply(lambda x : float(x[:-1]))
df_HCG["Time"] = df_HCG["Time"].apply(lambda x : float(x[:-1]))
df_SDG["Time"] = df_SDG["Time"].apply(lambda x : float(x[:-1]))
```

```python
plt.figure(figsize=[20,10])
sns.boxplot(x = df_SA["Instance"], y = df_SA["Time"], hue = df_SA["Move"]).set_title("Simulated Annealing - RS")

plt.figure(figsize=[20,10])
sns.boxplot(x = df_SD["Instance"], y = df_SD["Time"], hue = df_SD["Move"]).set_title("Steepest Descent - RS")

plt.figure(figsize=[20,10])
sns.boxplot(x = df_HC["Instance"], y = df_HC["Time"], hue = df_HC["Move"]).set_title("Hill Climbing - RS")

plt.figure(figsize=[20,10])
sns.boxplot(x = df_SAG["Instance"], y = df_SAG["Time"], hue = df_SAG["Move"]).set_title("Simulated Annealing - GS")

plt.figure(figsize=[20,10])
sns.boxplot(x = df_SDG["Instance"], y = df_SDG["Time"], hue = df_SDG["Move"]).set_title("Steepest Descent - GS")

plt.figure(figsize=[20,10])
sns.boxplot(x = df_HCG["Instance"], y = df_HCG["Time"], hue = df_HCG["Move"]).set_title("Hill Climbing - GS")
```

```python
#RANDOM INITIAL STATES
df_fin = pd.concat([df_SA[["Instance", "Move", "Profit"]], df_HC[["Instance", "Move", "Profit"]],  df_SD[["Instance", "Move", "Profit"]]], axis=0)

instances = ["p03", "p06", "p07", "p08", "p09", "p10", "p13", "p14", "p16"]
Greedy = [1239.58, 653.632, 1157.42, 1239.58, 1833.61, 2788.6, 1028.04, 1519.98, 2780.97]

for i,j in zip(instances, Greedy):
  df_fin_i = df_fin[df_fin["Instance"] == i]
  plt.figure(figsize=[20,10])
  plt.text(5, j + 10, "Greedy", size='medium', color='red')
  sns.boxplot(x = df_fin_i["Move"], y = df_fin_i["Profit"]).set_title("Instance " + i)
  sns.lineplot(y = j, x = df_fin_i["Move"].unique(), color = "red") 
```
